Q--10---Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small.
If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
## Define conditions based on 'total_bedrooms' values
conditions = [(df['total_bedrooms'] <=10),  ## Condition 1: If total bedrooms is 10 or less
              (df['total_bedrooms'] >=11)&(df['total_bedrooms'] <=1000), # Condition 2: If total bedrooms is between 11 and 1000
              (df['total_bedrooms']>1000)]  # Condition 3: If total bedrooms is greater than 1000

values = ['small','medium','large'] # Define corresponding values for each condition
## Create a new column 'total_bedroom_size' based on the conditions using np.select()
df['total_bedroom_size']=np.select(conditions,values)

df.sample(220)
```
